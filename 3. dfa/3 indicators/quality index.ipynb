```python
import pandas as pd
import numpy as np
import matplotlib as plt
```

```python
df = pd.read_csv('/Users/tanishq/PycharmProjects/NIScPR/FinTech/DerwentData_TS/dfa/MainData.csv')
df['Application Date'] = pd.to_datetime(df['Application Date'])
df = df[(df['Application Date'] >= '2023-01-01') & (df['Application Date'] <= '2024-12-31')]
```

```python
df['Quality Index'] = (
    0.691083 * df['Count of Cited Refs - Non-patent'] +
    0.293741 * df['Count of Cited Refs - Patent'] +
    0.015176 * df['Count of Citing Patents']
)
```

```python
df.head()
```

```python
df = df.sort_values(by="Quality Index", ascending=False)
df.head()
```

```python
df.to_csv('QualityIndex(23-24)_3.csv', index=False)
print('Done!')
```
