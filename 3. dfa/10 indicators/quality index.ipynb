```python
import pandas as pd
import numpy as np
import matplotlib as plt
```

```python
df = pd.read_csv('/Users/tanishq/PycharmProjects/NIScPR/FinTech/DerwentData_TS/MainData.csv')
df['Application Date'] = pd.to_datetime(df['Application Date'])
df = df[(df['Application Date'] >= '2023-01-01') & (df['Application Date'] <= '2024-12-31')]
```

```python
df['Quality Index'] = (
    0.180466 * df['Count of Cited Refs - Patent'] +
    0.420173 * df['Count of Cited Refs - Non-patent'] +
    0.011189 * df['Count of Citing Patents'] +
    0.058835 * df['DWPI Count of Family Members'] +
    0.114706 * df['DWPI Count of Family Countries/Regions'] +
    0.032466 * df['Assignee Count'] +
    0.053108 * df['Inventor Count'] +
    0.042724 * df['Claims Count'] +
    0.054720 * df['Legal Years Remaining'] +
    0.031613 * df['IPC Count']
)
```

```python
df = df.sort_values(by="Quality Index", ascending=False)
df.head()
```
