```python
import pandas as pd
import numpy as np
import matplotlib as plt
```

```python
df = pd.read_csv('/Users/tanishq/PycharmProjects/NIScPR/FinTech/DerwentData_TS/dfa/MainData.csv')
df['Application Date'] = pd.to_datetime(df['Application Date'])
df = df[(df['Application Date'] >= '2023-01-01') & (df['Application Date'] <= '2024-12-31')]
```

```python
df['Quality Index'] = (
    0.154116 * df['DWPI Count of Family Members'] +
    0.294948 * df['DWPI Count of Family Countries/Regions'] +
    0.083342 * df['Assignee Count'] +
    0.136223 * df['Inventor Count'] +
    0.109425 * df['Claims Count'] +
    0.140872 * df['Legal Years Remaining'] +
    0.081073 * df['IPC Count']
)
```

```python
df = df.sort_values(by="Quality Index", ascending=False)
df.head()
```
